```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import signal

fs = 128  # Sampling frequency
window_size = 100  # Number of samples per window

# Define EEG bands (excluding Gamma)
bands = {'Delta': (0.5, 4), 'Theta': (4, 8), 'Alpha': (8, 13), 'Beta': (13, 30)}

for x in df.columns[1:-1]:  # Iterating over EEG channels
    data = df[x].astype(float).values

    # Split the data into non-overlapping windows
    num_windows = len(data) // window_size
    data_windows = np.array_split(data[:num_windows * window_size], num_windows)

    # Initialize storage for band power
    band_powers = {band: [] for band in bands}

    # Process each window
    for window in data_windows:
        f, psd = signal.welch(window, fs, nperseg=len(window))  # Compute PSD

        for band, (low, high) in bands.items():
            band_power = np.trapz(psd[(f >= low) & (f <= high)], f[(f >= low) & (f <= high)])  # Integrate PSD
            band_powers[band].append(band_power)

    # Convert to DataFrame
    band_powers_df = pd.DataFrame(band_powers)

    # Plot the power in each band across time windows
    plt.figure(figsize=(12, 6))
    for band in bands.keys():
        plt.plot(band_powers_df.index, band_powers_df[band], label=band)

    plt.xlabel(f"Window of {x}")
    plt.ylabel('Power')
    plt.title(f'EEG Band Power for {x} over Time Windows')
    plt.legend()
    plt.grid()
    plt.show()
```
